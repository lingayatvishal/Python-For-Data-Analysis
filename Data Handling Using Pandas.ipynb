# Data cleaning

```python
df
```

```python
#A lot of '?' sign in the dataset. Since pandas don't consider this as missing data, we will have to convert this into Nan.
```

```python
df.replace("?",np.nan,inplace =True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
#Now we can see all the missing values in the dataset. Let's try to fill these data with mean values except for num-of-doors.
```

```python
#Since both the missing values in num-of-doors is of sedan, we will replace the Nan value with four
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
#Drop all the rows with missing price data
```

```python
df.dropna(subset =['price'], axis = 0, inplace =True)
```

```python
#Replace all the missing values with mean
avg_loss = df['normalized-losses'].astype('float').mean()
```

```python
avg_loss
```

```python
df['normalized-losses'].replace(np.nan, avg_loss, inplace = True)
```

```python
avg_bore = df['bore'].astype('float').mean()
avg_bore
```

```python
df['bore'].replace(np.nan, avg_bore, inplace = True)
```

```python
avg_stroke = df['stroke'].astype('float').mean()
avg_stroke
```

```python
df['stroke'].replace(np.nan, avg_stroke, inplace = True)
```

```python
avg_hp = df['horsepower'].astype('float').mean()
avg_hp
```

```python
df['horsepower'].replace(np.nan, avg_hp, inplace = True)
```

```python
avg_rpm = df['peak-rpm'].astype('float').mean()
avg_rpm
```

```python
df['peak-rpm'].replace(np.nan, avg_rpm, inplace = True)
```
